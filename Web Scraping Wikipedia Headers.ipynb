Till now we had obtained the list of First 50 movies,now lets take the webpage of next 50 movie lists. 

```python
#next 50 list of movies
page=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")

#parsing this code using BeautifulSoup
soup2=BeautifulSoup(page.content,'html.parser')
#converting this parsed structure into understandable format using prettify().
print(soup2.prettify())
```

```python
#extracting html tags of all movie names 
header_tags=soup2.find_all('h3',class_="lister-item-header")
header_tags[0:4]
```

```python
#extracting the text from these tags
for i in header_tags:
    Header_tags.append(i.get_text())
Header_tags[:]
```

```python
#extracting the html tags of ratings
ratings=soup2.find_all('strong')
ratings[:]
```

```python
#extracting the text from these tags
for i in ratings:
    Rating.append(i.get_text())
Rating[:]
```

```python
len(Rating)
```

```python
#removing the characters
Rating.remove('Detailed')
Rating.remove('User Rating')
```

```python
Rating
```

```python
len(Rating)
```

```python
#creating a dataframe with the list of top 100 IMDb rated movies
ds=pd.DataFrame()
ds['Headers']=Header_tags[:]
ds['Movie']=ds['Headers'].str.split('\n').str[2]
ds['Year']=ds['Headers'].str.split('\n').str[3]
ds['Rating']=Rating
ds.drop(columns='Headers',axis=1,inplace=True)
ds
```

```python
#saving the dataframe into a csv file
ds.to_csv('Top 100 movies by IMDb rating.')
```
