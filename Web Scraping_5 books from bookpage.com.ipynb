```python
#importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://bookpage.com/reviews?book_genre=fiction&page=1")
page
```

```python
#parsing this code using BeautifulSoup
soup=BeautifulSoup(page.content,'html.parser')
#converting this parsed structure into understandable format using prettify().
print(soup.prettify())
```

```python
#creating empty lists
Book_Name=[]
Author_Name=[]
Genre=[]
Review=[]
```

```python
#extracting the html tags for book names
bookname=soup.find_all('h4',class_="italic")
bookname[0:5]
```

```python
#extracting the text from these tags
for i in bookname:
    Book_Name.append(i.get_text())
Book_Name[0:5]
```

```python
#extracting the html tags for the authors name
authorname=soup.find_all('p',class_="sans bold")
authorname[0:5]
```

```python
#extracting the text from these tags
for i in authorname:
    Author_Name.append(i.get_text())
Author_Name[0:5]
```

```python
#extracting the html tags for the genre
genre=soup.find_all('p',class_="genre-links hidden-phone")
genre[0:5]
```

```python
#extracting the text from these tags
for i in genre:
    Genre.append(i.get_text())
Genre[0:5]
```

```python
#extracting the html tags for review
review=soup.find_all('p',class_="excerpt")
review[0:5]
```

```python
#extracting the text from these tags
for i in review:
    Review.append(i.get_text())
Review[0:5]
```

```python
#creating a dataframe with the list of 5 books
ds=pd.DataFrame()
ds['Book_Name']=Book_Name[0:5]
ds['Book_Name']=ds['Book_Name'].str.split('\n').str[1]
ds['Author_Name']=Author_Name[0:5]
ds['Author_Name']=ds['Author_Name'].str.split('\n').str[1]
ds['Genre']=Genre[0:5]
ds['Genre1']=ds['Genre'].str.split('\n').str[1]
ds['Genre2']=ds['Genre'].str.split('\n').str[3]
ds.drop(columns='Genre',axis=1,inplace=True)
ds['Review']=Review[0:5]
ds['Review']=ds['Review'].str.split('\n').str[1]
ds
```

```python
#saving the dataframe to csv file
ds.to_csv('Webscrap of Five books')
```
