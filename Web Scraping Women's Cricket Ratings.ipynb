```python
#importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Top 10 ODI Team Rankings

```python
page1=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
page1
```

```python
#parsing this code using BeautifulSoup
soup1=BeautifulSoup(page1.content,'html.parser')
#converting this parsed structure into understandable format using prettify().
print(soup1.prettify())
```

```python
#creating empty lists
Team=[]
Matches=[]
Points=[]
Ratings=[]
```

```python
#extracting the html tags for Team name
team=soup1.find_all('span',class_="u-hide-phablet")
team[0:10]
```

```python
#extracting the text from these tags
for i in team:
    Team.append(i.get_text())
Team[0:10]
```

```python
#extracting the html tags for Matches
matches=soup1.find_all('td',class_=["rankings-block__banner--matches","table-body__cell u-center-text"])
matches[:]
```

```python
#extracting the text from these tags
for i in matches:
    Matches.append(i.get_text())
Matches[:]
```

```python
#create a list to seperate points
list2=Matches[2:20:2]
list2
```

```python
# seperating matches from the  lists
matches1=[Matches[0]]
matches2=Matches[1:19:2]
```

```python
matches1.extend(matches2)
Matches=matches1
Matches
```

```python
#extracting the html tags for points
points=soup1.find('td',class_="rankings-block__banner--points")
points
```

```python
#extracting text from the tags
Points.append(points.get_text())
Points
```

```python
#appending the list of succesive team points
Points.extend(list2)
Points
```

```python
#extracting the html tags for rating
rating1=soup1.find('td',class_=["rankings-block__banner--rating u-text-right"])
Ratings=[rating1.text]
```

```python
Ratings
```

```python
rating2=soup1.find_all('td',class_=["table-body__cell u-text-right rating"])
rating2[:9]
```

```python
for i in rating2:
    Ratings.append(i.get_text())
Ratings[:]
```

```python
Ratings[0]='123'
Ratings=Ratings[:10]
Ratings
```

```python
ds1=pd.DataFrame()
ds1['Team']=Team[0:10]
ds1['Matches']=Matches[0:10]
ds1['Points']=Points[0:10]
ds1['Rating']=Ratings[0:10]
ds1
```

### Top 10 ODI Womens Batting rankings

```python
page2=requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi")
page2
```

```python
#parsing this code using BeautifulSoup
soup2=BeautifulSoup(page2.content,'html.parser')
#converting this parsed structure into understandable format using prettify().
print(soup2.prettify())
```

```python
#creating empty lists
Player=[]
Player1=[]
Team=[]
Team1=[]
Rating=[]
```

```python
#extracting html tags for Player name
player=soup2.find('div',class_="rankings-block__banner--name")
player
```

```python
#extracting text from the tags
Player.append(player.text)
Player
```

```python
#extracting the html tags for the remaining player names
player1=soup2.find_all('td',class_="table-body__cell name")
player1[0:9]
```

```python
#extracting text from these tags
for i in player1:
    Player1.append(i.get_text())
Player1[0:9]    
```

```python
#removing '\n' from the list
Player1=[s.strip('\n') for s in Player1]
Player1
```

```python
Player.extend(Player1[0:9])
Player
```

```python
#extracting the html tags for Team
team1=soup2.find('div',class_="rankings-block__banner--nationality")
team1
```

```python
#extracting the text from these tags
Team.append(team1.get_text())
Team
```

```python
#removing '\n' from the list
Team=[s.strip('\n') for s in Team]
Team
```

```python
#extracting html tags for successive teams
team2=soup2.find_all('span',class_="table-body__logo-text")
team2
```

```python
#extracting text from these tags
for i in team2:
    Team1.append(i.get_text())
Team1[0:9]    
```

```python
Team.extend(Team1[0:9])
Team
```

```python
#extracting the html tags for rating
rating=soup2.find('div',class_="rankings-block__banner--rating")
rating
```

```python
#extracting text from the tag
Rating.append(rating.text)
Rating
```

```python
#extracting the html tags for successive players ratings
rating1=soup2.find_all('td',class_="table-body__cell u-text-right rating")
rating1
```

```python
#extracting text from these tags
for i in rating1:
    Rating.append(i.get_text())
Rating[0:10]    
```

```python
#creating a dataframe
ds2=pd.DataFrame()
ds2['Player']=Player[0:10]
ds2['Team']=Team[0:10]
ds2['Rating']=Rating[0:10]
ds2
```
